```python
x_train_cat=x_train[['Gender', 'Married', 'Education',
       'Self_Employed', 'Property_Area','Dependents']]
```

```python
x_train_cat
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
for i in x_train_cat.columns:
    x_train[i]=lb.fit_transform(x_train[i])
```

```python
x_train_w_id=x_train.drop("Loan_ID",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train[["ApplicantIncome" ,"CoapplicantIncome" ,"LoanAmount","Loan_Amount_Term"]]=ss.fit_transform(x_train[["ApplicantIncome" ,"CoapplicantIncome" ,"LoanAmount","Loan_Amount_Term"]])
```

```python
x_train_w_id
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
score=cross_val_score(lr,x_train_w_id,y_train,scoring="neg_mean_squared_error",cv=10)
```

```python
y_train=lb.fit_transform(y_train)
```

```python
import numpy as np
rmse=np.sqrt(-score)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
score_1=cross_val_score(dtc,x_train_w_id,y_train,scoring="neg_mean_squared_error",cv=10)
```

```python
rmse=np.sqrt(-score_1)
```
